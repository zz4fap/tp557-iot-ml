### Neste exemplo, vamos explorar a função de erro, também conhecida como função de custo.

Import Libraries

```python
import matplotlib.pyplot as plt
```

Define input data and target data

```python
x = [-1, 0, 1, 2, 3, 4]
y = [-3, -1, 1, 3, 5, 7]
```

Plot data

```python
plt.scatter(x, y, color='red')
plt.axis([-1,5,-5,8])
plt.title('Data')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (rótulos)')
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', label='y')
plt.axis([-1,5,-5,8])
plt.title('Data')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (rótulos)')
plt.grid()
plt.show()
```

Edit parameters ($a_0$ and $a_1$) to try different loss measurements.

- Rerun the cells when done
- Your $\hat{y}$ will be calculated as: $\hat{y}  = a_0 + a_1 x$,

Example: if $a_0=-1$, and $a_1=3$, then: $y=3x-1$

```python
a0 = -1
a1 = 3
```

#### Calculate and plot y_hat

```python
y_hat = []

for xi in x:
    yi = a0 + (a1*xi)
    y_hat.append(yi)
    print(y_hat)

print("")
print("Real y is " + str(y))
print("My y is   " + str(y_hat))
```

```python
plt.plot(x, y, color='red', marker='o', label='y')
plt.plot(x, y_hat, color='blue', marker='o', label='predições')
plt.axis([-1,5,-5,15])
plt.title('Predições')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (saída)')
plt.legend(loc='upper left')
plt.grid()
```

Let's calculate the loss

```python
N = len(y)
```

```python
total_squared_error = 0
for i in range(0, N):
    squared_error = (y[i] - y_hat[i]) ** 2
    total_squared_error += squared_error
    print(total_squared_error)

print("")
print("My total loss is: {:.2f}".format(total_squared_error))
```

Let's calculate the MSE (Mean Squared Error) that is typical in Regression problems

```python
print("My MSE loss is: {:.2f}".format((1/N)*(total_squared_error)))
```

Criating a function

```python
def calc_plot_error (x, y, a0, a1):
    y_hat = []
    for xi in x:
        yi = a0 + (a1*xi)
        y_hat.append(yi)

    total_squared_error = 0
    N = len(y)
    for i in range(0, N):
        squared_error = (y[i] - y_hat[i]) ** 2
        total_squared_error += squared_error

    MSE = (1/N)*(total_squared_error)
    print("MSE is: {:.2f}".format(MSE))
    plt.plot(x, y, color='red', marker='o', label='y')
    plt.plot(x, y_hat, color='blue', marker='o', label='predictions')
    plt.axis([-1,5,-5,15])
    plt.title('Predições')
    plt.xlabel('x (dados de entrada)')
    plt.ylabel('y (saída)')
    #plt.legend(loc='upper left')
    plt.grid()

    return MSE
```

```python
calc_plot_error (x, y, a0=-2, a1=2)
```

Let's fix the parameter $a_0$ and study $a_1$

```python
a0 = -1
a1_lst = [-2, -1, 0, 1, 2, 3, 4, 5, 6]
MSE_lst = []
for a1i in a1_lst:
    MSEi = calc_plot_error (x, y, a0=a0, a1=a1i)
    MSE_lst.append(MSEi)
```

```python
MSE_lst
```

```python
plt.plot(a1_lst, MSE_lst, color='orange', marker='o', label='error')
plt.title('Função de erro')
plt.xlabel('$a_1$')
plt.ylabel('Erro')
plt.grid()
plt.show()
```
