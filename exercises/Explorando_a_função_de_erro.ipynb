## Explorando a função de erro

Neste exemplo, vamos explorar a função de erro, também conhecida como função de custo.

#### Importando as bibliotecas

```python
import matplotlib.pyplot as plt
```

#### Definindo os dados de entrada e os rótulos (i.e., valores de saída esperados)

```python
x = [-1, 0, 1, 2, 3, 4]
y = [-3, -1, 1, 3, 5, 7]
```

#### Plotando os dados gerados.

```python
plt.scatter(x, y, color='red')
plt.axis([-2, 5, -5, 8])
plt.title('Data')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (rótulos)')
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', label='y')
plt.axis([-2, 5, -5, 8])
plt.title('Data')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (rótulos)')
plt.grid()
plt.show()
```

#### Modificando os parâmetros ($a_0$ and $a_1$) para verificar medidas de erro diferentes.

- Reexecute a célula após a modificação
- O novo $\hat{y}$ será calculado como: $\hat{y}  = a_0 + a_1 x$,

Exemplo: Se $a_0=-1$, e $a_1=3$, então: $y=3x-1$

```python
a0 = -1
a1 = 3
```

#### Calculando e plotando y_hat

```python
y_hat = []

for xi in x:
    yi = a0 + a1*xi
    y_hat.append(yi)
    print(y_hat)

print("")
print("Real y is " + str(y))
print("My y is   " + str(y_hat))
```

```python
plt.plot(x, y, color='red', marker='o', label='y')
plt.plot(x, y_hat, color='blue', marker='o', label='predições')
plt.axis([-1,5,-5,15])
plt.title('Predições')
plt.xlabel('x (dados de entrada)')
plt.ylabel('y (saída)')
plt.legend(loc='upper left')
plt.grid()
```

#### Vamos calcular o erro

```python
N = len(y)
total_squared_error = 0
for i in range(0, N):
    squared_error = (y[i] - y_hat[i]) ** 2
    total_squared_error += squared_error
    print(total_squared_error)
    
MSE = (1/N)*total_squared_error

print("")
print("O MSE (Mean Squared Error) é: {:.2f}".format(MSE))
```

#### Criando uma função que calcula o MSE e plota a função hipótese

```python
def calc_plot_error (x, y, a0, a1):
    y_hat = []
    for xi in x:
        yi = a0 + a1*xi
        y_hat.append(yi)

    total_squared_error = 0
    N = len(y)
    for i in range(0, N):
        squared_error = (y[i] - y_hat[i]) ** 2
        total_squared_error += squared_error

    MSE = (1/N)*(total_squared_error)
    print("MSE is: {:.2f}".format(MSE))
    
    plt.plot(x, y, color='red', marker='o', label='Dados originais')
    plt.plot(x, y_hat, color='blue', marker='o', label='Função hipótese')
    plt.axis([-1,5,-5,15])
    plt.title('Predições')
    plt.xlabel('x (dados de entrada)')
    plt.ylabel('y (saída)')
    #plt.legend(loc='upper left')
    plt.grid()

    return MSE
```

```python
calc_plot_error (x, y, a0=-2, a1=2)
```

#### Vamos fixar o valor do peso $a_0$ e variar o peso $a_1$

```python
a0 = -1
a1_lst = [-2, -1, 0, 1, 2, 3, 4, 5, 6]
MSE_lst = []
for a1i in a1_lst:
    MSEi = calc_plot_error (x, y, a0=a0, a1=a1i)
    MSE_lst.append(MSEi)
```

```python
MSE_lst
```

```python
plt.plot(a1_lst, MSE_lst, color='orange', marker='o', label='error')
plt.title('Função de erro')
plt.xlabel('$a_1$')
plt.ylabel('Erro')
plt.grid()
plt.show()
```
