### Plotando o conjunto de dados.

```python
plt.plot(x, y, '.')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

### Definina a rede neural

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

### Imprima o resumo da arquitetura do modelo

```python
model.summary()
```

### Compile o modelo

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

### Treine o modelo

```python
# Ajusta o modelo aos dados (também conhecido como treinar o modelo)
history = model.fit(x, y, epochs=500)
```

### Plote o histórico de erros ao longo das épocas de treinamento

```python
type(history.history)
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.grid()
plt.show()
```

### Compare os dados ruidosos com as predições feitas pelo modelo.

```python
plt.plot(x, y, '.')
plt.plot(x, model.predict(x))
plt.grid()
plt.show()
```
